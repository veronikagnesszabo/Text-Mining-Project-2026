```python
from transformers import pipeline

# Initialize the sentiment analysis pipeline
sentiment_task = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")

# Test it with a tweet
sample_text = "I love horses, they are my favourite animal! "
result = sentiment_task(sample_text)

print(result)
# Output: [{'label': 'positive', 'score': 0.99}]

test_sentences = [
    "I'm so happy this finally works!",
    "This was a very frustrating error.",
    "Today is Tuesday. I love my work but I hate the weather. I am going home now"
]

results = sentiment_task(test_sentences)

for text, res in zip(test_sentences, results):
    print(f"Text: {text} | Label: {res['label']} | Score: {res['score']:.4f}")
```

```python

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


sentences = ["I am the best",
             "I AM THE BEST!",
             "Not bad at all",

             ]

analyzer = SentimentIntensityAnalyzer()
for sentence in sentences:
    vs = analyzer.polarity_scores(sentence)
    print("{:-<65} {}".format(sentence, str(vs)))
```
