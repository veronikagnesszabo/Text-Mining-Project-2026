```python
import pandas as pd
import nltk
import spacy
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')

df = pd.read_csv('/Users/camellia/Downloads/Text-Mining-Project-2026/ted_main.csv')
print(f" dataset has {len(df)} lines")
print(df.columns.tolist())
```

```python
# Take a sample of 50 talks and load spacy
nlp = spacy.load("en_core_web_sm")
sample = df.sample(50, random_state=42)

def preprocess(text):
    doc = nlp(str(text))
    tokens = [token.lemma_.lower() for token in doc 
              if not token.is_stop and not token.is_punct and not token.is_space]
    return tokens

# Apply to sample
sample = sample.copy()
sample['processed_tokens'] = sample['description'].apply(preprocess)
print("Preprocessing complete!")
print(sample[['name', 'processed_tokens']].head(3))
```

```python
# Save preprocessed sample to results folder
sample[['name', 'description', 'processed_tokens', 'tags', 'views', 'comments']].to_csv(
    '/Users/camellia/Downloads/Text-Mining-Project-2026/results/preprocessed_sample.csv',
    index=False
)
```

```python
# Apply preprocessing pipeline to all transcripts
df['clean_text'] = df['description'].apply(preprocess)

# Save to results folder
df[['name', 'description', 'clean_text', 'tags', 'views', 'comments']].to_csv(
    '/Users/camellia/Downloads/Text-Mining-Project-2026/results/preprocessed_full.csv',
    index=False
)
print("All transcripts preprocessed.")
```

```python
transcripts = pd.read_csv('/Users/camellia/Downloads/Text-Mining-Project-2026/transcripts.csv')
print(transcripts.columns.tolist())
print(transcripts.head(2))
```

```python
# Merge transcripts with main dataset on url
df_merged = df.merge(transcripts, on='url', how='inner')
print(f"Merged dataset has {len(df_merged)} rows")
print(df_merged[['name', 'transcript']].head(2))
```

```python
# Apply preprocessing to transcript column
df_merged['clean_tokens'] = df_merged['transcript'].apply(preprocess)
df_merged['clean_text_joined'] = df_merged['clean_tokens'].apply(lambda tokens: ' '.join(tokens))

# Save to results folder
df_merged[['name', 'transcript', 'clean_tokens', 'clean_text_joined', 'tags', 'views', 'comments']].to_csv(
    '/Users/camellia/Downloads/Text-Mining-Project-2026/results/preprocessed_full.csv',
    index=False
)
print("Done the preprocessing of transcript")
print(df_merged[['transcript', 'clean_text_joined']].head(2))
```
